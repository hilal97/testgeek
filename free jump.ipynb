```python
Name= "Michael Jackson"
len(Name)
#Indexing
print(Name[0])
print(Name[8])
print(Name[-1])
print(Name[13])
print(Name[-3])
print(Name[-15])
```

```python
# Slicing: We can obtain multiple characters from a string using slicing, we can obtain the 0 to 4th and 8th to the 12th element:  
Name[:4] #range from 0 to 3
```

```python
Name[1:9] #range from 0 to 8
```

```python
#Stride:  We can also input a stride value as follows, with the '2' indicating that we are selecting every second variable:
Name[::2] #jumping one row one after another and do not stop / Get every second element. The elments on index 1, 3, 5 ...
```

```python
# Get every second element in the range from index 0 to index 7
Name[0:8:2] #jumping one row one after another and stop at 7.
```

```python
Name[0:8:3] #jumping two rows one after another and stop at 7.
```

```python
#Slicing on a list
L=['Michael Jackson', 10.1, 1982, "M", 1]
L[1:4]
```

```python
#Append
#Another similar method is append If we apply append instead of extend, we add one element to the list.
L.append(['hard rock'])
L
```

```python
#deleting an element of a list.
#We can also delete an element of a list using the del command:
del(L[3])
L
```

```python
#Converting a string to a list
#We can convert a string to a list using split. For example, the method split translates every group of characters separated by a space into an element in a list:
'Johan Kouame'.split()
```
